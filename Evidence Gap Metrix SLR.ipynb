```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Evidence-gap data
data = {
    "Cognitive": [12, 3, 5],
    "Relational": [10, 2, 4],
    "Procedural": [9, 3, 5],
    "Institutional": [3, 0, 1]
}
```

```python
df = pd.DataFrame(data, index=['Tabletop exercises (TTX)', 'Digital / virtual simulations', 'Hybrid / large-scale simulations'])
```

```python
# Create heatmap
plt.figure(figsize=(10,5))

sns.heatmap(
    df,
    annot=True,
    #fmt="d",
    cmap="Blues",
    cbar_kws={'label': 'Number of Studies'},
    linewidths=1,
    #linecolor="white"
)

#plt.title("Evidence Gap Heatmap: Governance Outcomes of Simulation-Based Coordination Interventions")
plt.xlabel("Governance Capacity Dimension", labelpad=15)
plt.ylabel("Simulation modality")
plt.tight_layout()

plt.show()
```
