```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns
```

```python
# Evidence-gap data
data = {
    "Cognitive": [12, 3, 5],
    "Relational": [10, 2, 4],
    "Procedural": [9, 3, 5],
    "Institutional": [3, 0, 1]
}
```

```python
df = pd.DataFrame(data, index=['Tabletop exercises (TTX)', 'Digital / virtual simulations', 'Hybrid / large-scale simulations'])
```

```python
# Create heatmap
plt.figure(figsize=(10,5))

sns.heatmap(
    df,
    annot=True,
    #fmt="d",
    cmap="Blues",
    cbar_kws={'label': 'Number of studies'},
    linewidths=1,
    #linecolor="white"
)

#plt.title("Evidence Gap Heatmap: Governance Outcomes of Simulation-Based Coordination Interventions")
plt.xlabel("Governance capacity dimension", labelpad=15)
plt.ylabel("Simulation modality")
plt.tight_layout()

plt.show()
```

```python
# Years of the included studies
years = [
2016, 2018, 2002, 2019, 2013, 2018, 2025, 2024, 2019, 2023, 2009,
2025, 2025, 2012, 2018, 2021, 2015, 2016, 2018, 2010, 2012, 2025,
2012, 2015, 2007, 2001, 2001, 2003, 2018, 2014, 2006, 2003, 2012
]

# Count studies per year
year_counts = Counter(years)

# Sort data
x = sorted(year_counts.keys())
y = [year_counts[year] for year in x]

# Create figure
plt.figure(figsize=(10, 6))
plt.bar(x, y)

# Labels and title (sentence case)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of included studies", fontsize=12)
plt.title("Temporal distribution of included studies", fontsize=14)

# Formatting
plt.xticks(x, rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()

# Save figure
plt.savefig("temporal_distribution_bar_chart.png", dpi=300, bbox_inches="tight")

# Show plot
plt.show()
```
